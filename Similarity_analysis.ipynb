```python
import difflib
import sys

def dividing_text_by_sentences(filename):
  '''
  Function which divide text from a file into sentences. Each string represents one sentence.

  Parameters:
  filename (str): file with our text

  Returns:
  list: list of strings which contain sentences
  '''

  with open(filename, 'r') as f:
    new_file = f.read() # reading a file

  new_file = new_file.split('\n') # spliting file on paragraphs

  file_splitted_into_sentences = []

  for line in new_file: # starting a loop in order to add every sentence from a file to a list, line represents one line in a file
    k = 0 # variable which represents one character in a file
    a = 0
    while k < len(line):
      if line[k] == '.':
        file_splitted_into_sentences.append(line[a:k])
        a = k + 1 # adding variable named a to avoid spaces appearing after the dot in sentence
        k += 1
      else:
        k +=1
        continue
  
  return file_splitted_into_sentences



def analyze_similarity(file1, file2):
  '''
  Function which analyze similarity ratio between two files
  
  Parameters:
  file1, file2 (list): list of strings which contain sentences

  Returns:
  int: the value of similarity ratio between two texts
  '''


  sum_similarity = 0 
  # defined sum of overall similarity between every sentence

  for i in range(len(file1)): 
    for k in range(len(file2)):
      sum_similarity += difflib.SequenceMatcher(lambda x: x == ' ', file1[i], file2[k]).ratio() # summing the similarity of each sentence from the first file with each sentence from the second file

  similarity = sum_similarity / (len(file1) * len(file2)) # overall similiarity divided by lenght of the first file multiplicated by the length of the second file so that there is no bias caused by sentence sizes in the files

  return similarity


def analyze_similarity_between_texts(filename1, filename2):
  '''
  Function which divide text from a file into sentences. 
  Each string represents one sentence. After that function analyzes similarity ratio between two files

  Parameters:
  filename1 (str): file with our text
  filename2 (str): file with our text

  Returns:
  int: the value of similarity ratio between two texts
  '''
  file1 = dividing_text_by_sentences(filename1) # dividing text into sentences from a file
  file2 = dividing_text_by_sentences(filename2) # dividing text into sentences from a file
  return analyze_similarity(file1, file2) # analyzing similarity ratio between two texts
```
